```python
import cv2
```

```python
img=cv2.imread(r'C:\Computer vision\Face_detection\20200917_170705.jpg')
img
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
con_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(con_img)
```

```python
img_g=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img_g)
```

```python
cascade=cv2.CascadeClassifier(r'C:\Computer vision\Face_detection\haarcascade_frontalface_default.xml')
```

```python
cascade
```

```python
face=cascade.detectMultiScale(img_g,scaleFactor=1.1,minNeighbors=6)
```

```python
face
```

```python
len(face)
```

```python
for(x,y,w,h) in face:
    cv2.rectangle(con_img,(x,y),(x+w,y+h),color=(200,3,0),thickness=3)
```

```python
plt.imshow(con_img)
```

```python
video=cv2.VideoCapture(r'C:\Computer vision\Face_detection\video.mp4')
while True:
    suc,frame=video.read()
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=cascade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=10)
    #print(len(faces))
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),3)
    cv2.imshow('Video',frame)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
