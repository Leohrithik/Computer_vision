```python
cv2.imshow("sfjas",img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
txt=pytesseract.image_to_string(img)
print(txt)
```

```python
data=pytesseract.image_to_data(img,output_type=pytesseract.Output.DICT)
data
```

```python
data['text']
```

```python
n=len(data['text'])
for i in range(n):
    if data['conf'][i]>70:
        (x,y,w,h)=data['left'][i],data['top'][i],data['width'][i],data['height'][i]
        print(x,y,w,h)
        img_box=cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),1)
cv2.imshow("sfjas",img_box)
cv2.waitKey()
cv2.destroyAllWindows()
```
